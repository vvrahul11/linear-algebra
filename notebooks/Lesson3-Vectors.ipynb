# Vector-Vector Multiplication: The dot product
---

```python
# Many ways to compute the dot product

v1 = np.array([1, 2, 3, 4, 5, 6])
v2 = np.array([0, -4, -3, 6, 5, 1])

# Method 1
dp0 = sum(np.multiply(v1, v2))
print(dp0)

# Method 4
dp4 = 0
# Loop over elments
for i in range(0, len(v1)):
    # multiply corresponsing element and sum
    dp4 = dp4 + v1[i] * v2[i]
print(dp4)
```

```python
# Method 2
dp1 = np.dot(v1, v2)
print(dp1)

# Method 3
dp3 = np.matmul(v1, v2)
print(dp3)
```

# Dot product properties
Dot product between two vectors is the elementwise multiplication followed by sum

1. a * (b+c) = a*b + a *c
2. a * (b*c) = (a*b) * c

---

```python
# Distributive property
# create random vectors

n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# The two results 
res1 = np.dot(a, (b +c))

res2 = np.dot(a , b) + np.dot(a, c) 

# compare them
print([res1, res2])
```

```python
# Associative property
# create random vectors
'''
When you want to take the dot poduct of two vectors v1 and v2 and further
another dotproduct with the output of v1 and v2 with v3. This is how it works

dot(dot(v1, v2), v3 )
dot(v1, v2) = x # First take dot product of v1 and v2
dot(x, v3) = y # Do multiplication of elements of v3 with output of v1 and v2 and do not sum

An experiment to prove this can be seen in the below cell
'''


n = 3
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# The two results
res1 = np.dot(a, np.dot(b, c))
res2 = np.dot(np.dot(a, b), c)

# compare them
print(res1)
print(res2)
```

# The dot product from a geometric perspective
---

The dot product of two vectors is the magnitude of two vectors scaled by the angel betweeen them.

alpha = |a| |b| cos(theta a, b)

theta a, b= acos(alpha/|a||b|)


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# two vectors
v1 = np.array([2, 4, -3])
v2 = np.array([0, -3, -3])

# compute angle between two vectors
ang = np.arccos(np.dot(v1, v2)/ np.linalg.norm(v1 * np.linalg.norm(v2)))

# draw them
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'r')

plt.axis((-6, 6, -6, 6))
plt.title('Angle between vectors: %s rad.' %ang)
plt.show()
```

# Check orthogonality of vectors

```python
'''
Two vectors are orthogonal if the angle between them is 90 degree. Which means if the
dot product between them is equal to 0. Cos(90) = 0
'''
v1 = np.array([ 16, -2, 4 ])
v2 = np.array([ .5, 2, -1 ])

# compute angle between two vectors
ang = np.dot(v1, v2)
ang

# draw them
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'r')

plt.axis((-10, 10, -10, 10))
plt.title('Angle between vectors: %s rad.' %ang)
plt.show()
```
